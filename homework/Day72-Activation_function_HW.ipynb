# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''
```

```python
def ReLU(x):
    return [max(0, xi) for xi in x]

def dReLU(x):
    dx = []
    for xi in x:           
        if xi < 0:
            dx.append(0)
        elif xi == 0:
            dx.append(np.nan)
        else:
            dx.append(1)
    return dx
```

```python
x = plt.linspace(-10,10,101)
x, dReLU(x)
plt.plot(x, ReLU(x), 'b', label='linspace(-10,10,21)')
plt.plot(x, dReLU(x), 'r', label='linspace(-10,10,21)')
plt.grid()
plt.title("ReLU Function")
plt.text(4, 1, r'$ReLU(x)=max(0,x)$', fontsize=15)

#resize the X and Y axes
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(0.5))
 
# create the graph
plt.show()
```
