```python
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline

n_clusters = [3,5]
init = ['k-means++', 'random']

for n in n_clusters:
    for i in init:

        km = KMeans(n_clusters=n, init=i)
        km.fit(X)
        km.labels_
        plt.scatter(X[:,2], X[:,3], c=km.labels_)
        plt.show()
        
        print('n_clusters={} init={}'.format(n,i))
        print(km.labels_)        
```
