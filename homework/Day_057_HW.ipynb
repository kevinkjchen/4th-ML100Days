- 作業目標 : 用 iris (dataset.load_iris()) 資料嘗試跑 hierarchical clustering

```python
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

載入 相關套件 並 執行 hierarchical clustering 實驗 ...

```python
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


linkage = ["ward", "complete", "average", "single"]

for lnkg in linkage:
    agg = AgglomerativeClustering(n_clusters=3, linkage=lnkg)
    agg.fit(X)
    labels = agg.labels_
    
    plt.scatter(X[:,2], X[:,3], c=agg.labels_)
    plt.show()
        
    print('linkage={}'.format(lnkg))
    print(agg.labels_)            
```

```python
# Plot the ground truth
petal_len_mean = []
petal_wth_mean = []
for name, label in [('cls0', 0),
                    ('cls1', 1),
                    ('cls2', 2)]:
    plt.scatter(X[y==label, 2].mean(), X[y==label, 3].mean(), marker='*', s=500, label=name)
    

    
# Reorder the labels to have colors matching the cluster results
cluster_labels = np.choose(y, [1, 0, 2])
plt.scatter(X[:,2], X[:,3], c=cluster_labels)
plt.legend()

print("cluster_labels={}".format(cluster_labels))
print("y={}".format(y))    
```
